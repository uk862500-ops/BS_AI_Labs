```python
##Task 1 – Even or Odd
num = int(input("enter a number"))
##Check if number is even or odd
if num%2 == 0:
    print(f"{num}, is even")
else:
    print(f"{num}, is odd")
```

```python
##Task 2 – Grade System

marks = int(input("enter your marks"))

if marks >= 80:
    print(f" your marks = {marks}, your grade is A")
elif marks >=70:
    print(f" your marks = {marks}, your grade is B")
elif marks >= 60:
    print(f"your marks = {marks}, your grade is C")
else:
    print(f" your marks is {marks}, you are fail")
```
